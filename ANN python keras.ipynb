```python
#http://machinelearningmastery.com/tutorial-first-neural-network-python-keras/

from keras.models import Sequential #Modelo secuencial, varias capas seguidas
from keras.layers import Dense # Capa fully connected
import numpy
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
#1. Cargar los datos

data = numpy.loadtxt("data/pima-indians-diabetes.data.txt", delimiter=",")

X = data[:,0:8]
Y = data[:,8]
```

```python
#Definir el modelo
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
#Modelo con 3 capas, la 1ra con 12 neuronas, 8 entradas y activacion relu
# Los pesos de las neuronas se inicializan con una distribucion uniforme, entre 0 y 0.5(por defecto en keras)
# La ultima capa usa sigmoid para tener una unica neurona con valor de 1 o 0.x
```

```python
#Compilar el modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Entrenamos el modelo para encontrar los mejores pesos
#Especificamos la f(x) de evaluacion de pesos (logarithmic loss o binary_crossentropy),
# el optimizador (SGD = adam) para buscar distintos pesos para la red y la metrica a usar
```

```python
#Entrenar el modelo - Ejecutarlo sobre datos
model.fit(X, Y, epochs=200, batch_size=50, verbose=2)
# Se ejecutara X iteraciones o epochs, cada 50 instancias se actualizan los pesos
```

```python
# Evaluar el modelo - miraremos el accuracy del entrenamiento, aunque no sabremos que tal lo hara sobre info nueva
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# Calcular predicciones, nos devolvera para cada dato una prediccion
predictions = model.predict(X)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
